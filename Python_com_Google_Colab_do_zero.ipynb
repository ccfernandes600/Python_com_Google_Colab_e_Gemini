No Google Colab, você pode usar uma variedade de comandos de terminal Linux para gerenciar arquivos, instalar pacotes e executar outras tarefas do sistema operacional. Para usar um comando de terminal no Colab, basta prefixá-lo com um ponto de exclamação (!).

### Comandos Básicos de Navegação e Manipulação de Arquivos no Google Colab

| `!rm arquivo`   | Remove um arquivo.                                |
| `!rm -r diretório` | Remove um diretório e todo o seu conteúdo (use com cuidado!). |
| `!mv origem destino`  | Move ou renomeia um arquivo ou diretório.        |
| `!cp origem destino`  | Copia um arquivo ou diretório.                    |

## Capítulo 4 - Noções básicas de Python (parte1)

Python é uma linguagem de programação multiparadigma de alto nível e tipada dinamicamente. O código Python é frequentemente considerado quase como um pseudocódigo, pois permite que você expresse ideias muito poderosas em pouquíssimas linhas de código, sendo muito legível. Como exemplo, aqui está uma implementação do calculo da Hipotenusa usando o teorema de pitágoras em Python:



````python
# Importa a biblioteca math para usar a função de raiz quadrada
import math

# Solicita ao usuário que insira os valores dos catetos
cateto_a = float(input("Digite o valor do primeiro cateto: "))
cateto_b = float(input("Digite o valor do segundo cateto: "))

# Calcula o quadrado de cada cateto
quadrado_a = cateto_a ** 2
quadrado_b = cateto_b ** 2

# Soma os quadrados dos catetos
soma_quadrados = quadrado_a + quadrado_b

# Calcula a raiz quadrada da soma dos quadrados para obter a hipotenusa
hipotenusa = math.sqrt(soma_quadrados)

# Exibe o resultado formatado
print(f"O valor da hipotenusa é: {hipotenusa:.2f}")
````

### **Tipos básicos de dados**

#### **Números Inteiros e Ponto flutuante**

**Inteiros (```int```)**:

* Representam números inteiros, positivos ou negativos, sem parte decimal.
* Exemplos: -5, 0, 10, 123456789.
* Usados para contar objetos, representar quantidades exatas, etc.

**Ponto Flutuante (```float```)**:

Representam números reais, com parte decimal.
Exemplos: 3.14, -0.5, 1e6 (notação científica para 1 milhão).
Usados para representar medidas, valores contínuos, resultados de cálculos, etc.

**Observações**:

* Python automaticamente determina o tipo de um número com base em sua representação.
* Cálculos envolvendo int e float geralmente resultam em float.
* O tipo float tem precisão limitada e pode apresentar pequenas imprecisões em cálculos.
* Para cálculos com precisão decimal arbitrária, use o módulo decimal.

Em Python, as estruturas de decisão permitem que seu programa escolha diferentes caminhos de execução com base em condições específicas. A principal ferramenta para isso é a instrução ```if```, que pode ser combinada com ```elif``` (```else if```) e ```else``` para criar lógicas mais complexas.



#### **```if```: A base das decisões**

A estrutura if avalia uma condição e executa um bloco de código somente se essa condição for verdadeira.

```python
idade= input(int("Entre com sua idade: "))
if idade >= 18:
```

#### **```else```: O caminho alternativo**

A cláusula ```else``` é usada em conjunto com o ```if``` para fornecer um bloco de código alternativo que será executado se a condição do ```if``` for falsa.

```python
idade= input(int("Entre com sua idade: "))
if idade >= 18:
    print("Você é maior de idade.")
else:
    print("Você é menor de idade.")
```

Agora, o programa exibirá "Você é menor de idade" se a idade for menor que 18.

```python
nota = 85
```

###  **Estruturas de Loop em Python: Repetindo tarefas com ```for``` e ```while```**

As estruturas de loop permitem que você execute um bloco de código repetidamente, seja por um número definido de vezes (```for```) ou enquanto uma condição for verdadeira (```while```).

#### **```for```: Iterando sobre sequências**

O loop ```for``` é ideal para percorrer sequências de dados, como listas, strings e intervalos numéricos.



```python
frutas = ["maçã", "banana", "laranja"]
```

Neste exemplo, o loop percorre a lista frutas e exibe cada elemento individualmente.


#### **```range()```: Gerando sequências numéricas**

A função ```range()``` é frequentemente usada com o ```for``` para gerar sequências de números:

```python
for i in range(5):  # De 0 a 4
```

```python
contador = 0
```

```python
for i in range(10):

    print(i)

for i in range(10):
    if i % 2 == 0:
        continue  # Pula os números pares
    print(i)
```

### Conjuntos (Sets) em Python: Coleções Desordenadas e Sem Duplicatas

Conjuntos (```sets```) em Python são coleções não ordenadas de elementos únicos. Isso significa que você não pode ter elementos repetidos dentro de um conjunto e que a ordem dos elementos não é garantida. Conjuntos são particularmente úteis para realizar operações matemáticas sobre conjuntos, como união, interseção e diferença.

####**Criando conjuntos:**

```python
numeros = {1, 2, 3, 4, 5}
```

####**Operações com conjuntos:**

```python
len(numeros)                # Saída: 6 (comprimento do conjunto)
numeros | {6, 7, 8}        # Saída: {1, 2, 3, 4, 5, 6, 7, 8} (união)
numeros & {4, 5, 6, 7}     # Saída: {4, 5, 6} (interseção)
numeros - {4, 5, 6, 7}     # Saída: {1, 2, 3} (diferença)
numeros ^ {4, 5, 6, 7}     # Saída: {1, 2, 3, 7, 8} (diferença simétrica)
3 in numeros                # Saída: True (verifica se 3 está no conjunto)
```

####**Percorrendo conjuntos com loops:**

```python
for numero in numeros:

```

####**Quando usar conjuntos:**

Para eliminar duplicatas de uma coleção de dados.
Para realizar operações matemáticas de conjuntos (união, interseção, etc.).
Para verificar a existência de um elemento em uma coleção de forma eficiente.

###Tratamento de Exceções em Python: Lidando com Erros como um Profissional

Em Python, exceções são eventos que interrompem o fluxo normal do seu programa quando algo inesperado acontece. Isso pode ser uma divisão por zero, um arquivo que não existe, um erro de digitação do usuário, entre outros. Felizmente, Python oferece um mecanismo poderoso para lidar com essas situações: o tratamento de exceções.

####**O que são exceções?**

Pense em exceções como alertas que seu programa emite quando encontra um problema. Em vez de travar abruptamente, o Python lança uma exceção, que você pode capturar e tratar de forma adequada.

####**Blocos** ```try-except```:

A estrutura básica para lidar com exceções é o bloco ```try-except```:

```python
try:
    # Código que pode gerar uma exceção
    resultado = 10 / 0
except ZeroDivisionError:
    # Código para lidar com a exceção ZeroDivisionError
    print("Erro: Divisão por zero!")
```

Neste exemplo:

1. O código dentro do bloco ```try``` é executado.
2.  Se ocorrer uma exceção ```ZeroDivisionError``` (divisão por zero), o programa pula para o bloco ```except``` correspondente.
3. O bloco ```except``` imprime uma mensagem de erro amigável.

####**Lidando com múltiplas exceções:**

Você pode ter vários blocos ```except``` para lidar com diferentes tipos de exceções:

```python
try:
    num = int(input("Digite um número: "))
    resultado = 10 / num
except ZeroDivisionError:
    print("Erro: Divisão por zero!")
except ValueError:
    print("Erro: Entrada inválida! Digite um número.")
```


####**Cláusulas** ```else``` e ```finally```:

* ```else```: Executa um bloco de código se nenhuma exceção ocorrer no bloco ```try```.
* ```finally``` : Executa um bloco de código sempre, independentemente de ocorrerem exceções.
```python
try:
    arquivo = open("dados.txt", "r")
    # ... processar o arquivo
except FileNotFoundError:
    print("Erro: Arquivo não encontrado!")
else:
    print("Arquivo processado com sucesso!")
finally:
    arquivo.close()  # Garante que o arquivo seja fechado
```

####**Lançando exceções:**

Você pode lançar suas próprias exceções usando a instrução ```raise```:

```python
def calcular_media(notas):
    if not notas:
        raise ValueError("A lista de notas está vazia!")
    # ... calcular a média
```

####**Exceções personalizadas:**

Você pode criar suas próprias classes de exceção para representar erros específicos do seu programa.

####**Boas práticas:**

* **Seja específico**: Capture exceções específicas em vez de usar um bloco ```except``` genérico.
* **Forneça mensagens claras**: Explique o erro ao usuário de forma clara e concisa.
* **Use** ```finally``` **para limpeza**: Garanta que recursos como arquivos sejam fechados, mesmo em caso de erros.
* **Não ignore exceções silenciosamente**: Lidar com as exceções ajuda a evitar problemas futuros.

###**Como usar o Gemini fora do Colab:**

1. **Acesse a API do Gemini**: O Google oferece uma API para acessar o Gemini programaticamente. Você precisará se inscrever e obter uma chave de API para usar o serviço.

2. **Instale a biblioteca cliente**: O Google fornece bibliotecas cliente para Python que facilitam a interação com a API do Gemini Pro. Você pode instalar a biblioteca usando o pip:

```bash
pip install google-cloud-aiplatform
Use o código com cuidado.
```
3. **Autenticação**: Configure a autenticação para usar sua chave de API e acessar o serviço.

4. Faça chamadas à API: Use as funções da biblioteca cliente para enviar solicitações ao Gemini e receber as respostas.

###**Exemplo de código (simplificado):**

```python
from google.cloud import aiplatform

# Configuração da API (substitua pelos seus valores)
aiplatform.init(project="SEU_PROJETO", location="SUA_LOCALIZAÇÃO")

# Crie um cliente para o Gemini Pro
client_options = {"api_endpoint": "us-central1-aiplatform.googleapis.com"}
client = aiplatform.gapic.PredictionServiceClient(client_options=client_options)

# Defina a instância do modelo
instance = {
    "prompt": "Qual é a capital da França?"
}

# Faça a solicitação à API
response = client.predict(
    endpoint="ENDPOINT_DO_SEU_MODELO",
    instances=[instance]
)

# Obtenha a resposta
resposta_gemini = response.predictions[0]["content"]
print(resposta_gemini)  # Saída: Paris
```
